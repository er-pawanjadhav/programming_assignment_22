```python
def simen_say(a,b):
    if len(a)==len(b) :
        if len(a)>=2:
            if a[0]==b[1]:
                return True
            else:
                return False
        else:
            return "lenght should be minimum 2"
    else:
        return "length of both list should be equal"
a = [1, 2, 3, 4, 5]
b = [5, 5, 1, 2, 3]

print(simen_say(a,b))
```
